## Code testing: practices, tools, and coverage

In the realm of research software engineering, the concepts of software testing and scientific reproducibility are inextricably linked. Traditional software development uses tests to ensure that code meets functional requirements and prevents regressions. In a scientific context, however, automated testing is the bedrock of reproducibility. It ensures that analytical pipelines, mathematical models, and data transformations reliably produce the exact same results across different environments and over time. Without a robust testing strategy, reproducibility remains a fragile and manual effort.

### Testing frameworks: `pytest` and alternatives

The Python ecosystem provides several excellent tools to implement automated testing. While the built-in `unittest` module offers a solid, object-oriented approach (heavily inspired by Java's JUnit), **`pytest`** has emerged as the *de facto* standard in the Python data science and engineering community. `pytest` is highly favoured for its concise, boilerplate-free syntax, its powerful `fixture` system for managing test states, and its detailed introspection of failing `assert` statements. Other alternatives, such as `nose2`, exist, but `pytest` provides the most modern, extensible, and readable foundation for testing scientific code.

### The Arrange-Act-Assert (AAA) pattern

Writing effective and maintainable tests requires structure. The **Arrange-Act-Assert (AAA)** pattern is an industry best practice that makes test suites clean and easy to understand:

* **Arrange**: Set up the necessary preconditions and inputs. In scientific software, this often involves generating synthetic datasets, initializing models, or configuring specific numerical parameters.
* **Act**: Execute the specific function, method, or analytical component being tested.
* **Assert**: Verify that the outcomes match the scientific expectations. This could mean checking if a calculated statistical value falls within an acceptable mathematical tolerance (`pytest.approx`) or ensuring a matrix maintains the correct dimensions.



To illustrate this pattern in practice, consider a test case for a function that calculates kinetic energy ($E_k = \frac{1}{2}mv^2$):

```python
import pytest

def calculate_kinetic_energy(mass: float, velocity: float) -> float:
    if mass < 0:
        raise ValueError("Mass cannot be negative")
    return 0.5 * mass * (velocity ** 2)

def test_calculate_kinetic_energy():
    # Arrange: Set up the initial physical parameters and expected outcome
    mass = 2.0          # in kilograms
    velocity = 3.0      # in meters per second
    expected_joules = 9.0

    # Act: Execute the function under test
    actual_joules = calculate_kinetic_energy(mass, velocity)

    # Assert: Verify that the numerical result matches the physical law
    assert actual_joules == pytest.approx(expected_joules)
```

### Code coverage and the 100% fallacy

To quantify testing efforts, developers rely on **code coverage** tools (such as the `coverage.py` package) and continuous integration services (like Codecov or Coveralls). These tools calculate the percentage of source code lines executed while the test suite runs, providing a useful metric for identifying untested analytical paths.

However, pursuing **100% code coverage** as an ultimate metric of quality is a dangerous fallacy. It is entirely possible to achieve complete coverage with a single, monolithic test that simply executes an entire pipeline from start to finish without asserting anything meaningful. Such a test proves only that the code runs without throwing fatal errors—it does absolutely nothing to prove that the outputs are mathematically or scientifically correct. 

The true goal of testing in research software is not merely to touch every line of code, but to build a comprehensive suite of granular assertions that validate the scientific logic, handle edge cases, and guarantee the numerical stability of the results.

## Why vector graphics are the only appropriate choice for non-photographic figures

Vector graphics are the preferred format for plots, diagrams, maps, charts, and most visualisations 
  in technical and professional documents.
Unlike raster graphics (e.g., PNG or JPG), which are composed of pixels, vector graphics (e.g., SVG or PDF) describe 
  shapes mathematically and represent text as structured elements.
As a result, they are resolution-independent and can be scaled without loss of quality.

Screenshots and raster exports (e.g., PNG or JPG) should be avoided for figures whenever 
  a vector format is available.
They should only be used for snapshots of software interfaces or content that cannot reasonably 
  be represented in vector form. 
Raster formats are primarily suitable for photographic images or other inherently pixel-based data.
Even for raster images, all text (labels, titles, annotations) should be added as vector graphics
  by embedding the images in SVG or PDF files which may contain a mixture of both vector and raster content. 
Raster exports of plots, maps and diagrams introduce loss of scalability, non-selectable text, 
  and reduced clarity when printed or zoomed.
  
Vector graphics are the expected format in academic journals and a best practice for professional materials 
  across industries, including logo design and visual identity systems.
SVG is a widely supported vector format viewable in all modern web browsers and editable using open-source 
  tools such as Inkscape, making it highly portable and interoperable.

````python
# | label: fig-sin
# | fig-cap: "Sinusoid & co."

import numpy as np
from matplotlib import pyplot

pyplot.plot(
    x := np.linspace(0, 2 * np.pi, 100), np.sin(x), label=r"$\sin(x)$"
)
pyplot.plot(x, np.cos(x / 2), label=r"$\cos(x/2)$")
pyplot.plot(
    x, np.sin(2 * x) * np.cos(x), label=r"$\sin(2x) \cdot \cos(x)$"
)
pyplot.legend();
````
