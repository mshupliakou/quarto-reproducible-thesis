## Code testing: practices, tools, and coverage

In the realm of research software engineering, the concepts of software testing and scientific reproducibility are inextricably linked. Traditional software development uses tests to ensure that code meets functional requirements and prevents regressions. In a scientific context, however, automated testing is the bedrock of reproducibility. It ensures that analytical pipelines, mathematical models, and data transformations reliably produce the exact same results across different environments and over time. Without a robust testing strategy, reproducibility remains a fragile and manual effort.

### Testing frameworks: `pytest` and alternatives

The Python ecosystem provides several excellent tools to implement automated testing. While the built-in `unittest` module offers a solid, object-oriented approach (heavily inspired by Java's JUnit), **`pytest`** has emerged as the *de facto* standard in the Python data science and engineering community. `pytest` is highly favoured for its concise, boilerplate-free syntax, its powerful `fixture` system for managing test states, and its detailed introspection of failing `assert` statements. Other alternatives, such as `nose2`, exist, but `pytest` provides the most modern, extensible, and readable foundation for testing scientific code.

### The Arrange-Act-Assert (AAA) pattern

Writing effective and maintainable tests requires structure. The **Arrange-Act-Assert (AAA)** pattern is an industry best practice that makes test suites clean and easy to understand:

* **Arrange**: Set up the necessary preconditions and inputs. In scientific software, this often involves generating synthetic datasets, initializing models, or configuring specific numerical parameters.
* **Act**: Execute the specific function, method, or analytical component being tested.
* **Assert**: Verify that the outcomes match the scientific expectations. This could mean checking if a calculated statistical value falls within an acceptable mathematical tolerance (`pytest.approx`) or ensuring a matrix maintains the correct dimensions.



To illustrate this pattern in practice, consider a test case for a function that calculates kinetic energy ($E_k = \frac{1}{2}mv^2$):

```python
import pytest

def calculate_kinetic_energy(mass: float, velocity: float) -> float:
    if mass < 0:
        raise ValueError("Mass cannot be negative")
    return 0.5 * mass * (velocity ** 2)

def test_calculate_kinetic_energy():
    # Arrange: Set up the initial physical parameters and expected outcome
    mass = 2.0          # in kilograms
    velocity = 3.0      # in meters per second
    expected_joules = 9.0

    # Act: Execute the function under test
    actual_joules = calculate_kinetic_energy(mass, velocity)

    # Assert: Verify that the numerical result matches the physical law
    assert actual_joules == pytest.approx(expected_joules)
```

### Code coverage and the 100% fallacy

To quantify testing efforts, developers rely on **code coverage** tools (such as the `coverage.py` package) and continuous integration services (like Codecov or Coveralls). These tools calculate the percentage of source code lines executed while the test suite runs, providing a useful metric for identifying untested analytical paths.

However, pursuing **100% code coverage** as an ultimate metric of quality is a dangerous fallacy. It is entirely possible to achieve complete coverage with a single, monolithic test that simply executes an entire pipeline from start to finish without asserting anything meaningful. Such a test proves only that the code runs without throwing fatal errors—it does absolutely nothing to prove that the outputs are mathematically or scientifically correct. 

The true goal of testing in research software is not merely to touch every line of code, but to build a comprehensive suite of granular assertions that validate the scientific logic, handle edge cases, and guarantee the numerical stability of the results.

### Academic precision: The MIT guidelines

Writing math in a thesis is more than making it readable. Following the **MIT Mathematical Typesetting Guidelines** here are some basic tips:

* **Variables** (e.g., $x, y, m$) represent changing values and must always be **italicized** (LaTeX does this by default in math mode).
* **Constants, operators, and differentials** (e.g., Euler's number $\mathrm{e}$, differential $\mathrm{d}x$) are fixed and must be written in **upright (Roman) font**. We achieve this using the `\mathrm{}` command.
* **Standard functions** (e.g., $\sin, \exp, \ln$) should also be upright. Use built-in LaTeX commands with a backslash (`\sin`) instead of typing plain text. This applies both the correct upright font and proper spacing.

> **Tip: Fine-tuning - spacing:** Notice the `\,` command in the integrals example above. It adds a small space before the differential, which is highly recommended.

Moreover, it is important to remember that mathematical exposition consists of sentences. The usage of good grammar and punctuation does not change. For stylistic reasons avoid beginning sentences with symbols.

### Automating \techword{.bib} file entries with Python

Managing bibliographies manually is error-prone and tedious.
Python provides two complementary tools for automating the process:

- **DOI content negotiation**: sending an HTTP `GET` to `https://doi.org/<doi>` with the header
  `Accept: application/x-bibtex` returns a ready-to-use BibTeX entry directly from the publisher.
  This works for virtually any paper that has a DOI, avoiding manual transcription errors.
- **`bibtexparser`**: a Python library for parsing, inspecting, and programmatically editing `.bib` files --
  useful for auditing large bibliographies, finding missing fields, or checking that all DOI links resolve.

In this handbook `refs.bib` is generated in a code cell at the top of the notebook.
The two cells below demonstrate both tools.

`````python
import requests


def doi_to_bibtex(doi):
    """Fetch a BibTeX entry for a given DOI via content negotiation."""
    response = requests.get(
        f"https://doi.org/{doi}",
        headers={"Accept": "application/x-bibtex"},
        allow_redirects=True,
        timeout=10,
    )
    response.raise_for_status()
    return response.text


# Example: the Matplotlib paper (Hunter, 2007)
try:
    print(doi_to_bibtex("10.1109/MCSE.2007.55"))
except requests.RequestException as exc:
    print(f"ERROR: {exc}")
`````

`````python
import bibtexparser


def validate_bib_entry(bib_entry):
    """Check that the entry's URL or DOI resolves correctly."""
    entry_doi = bib_entry.get("doi", None)
    url = bib_entry.get("url", None)
    try:
        if entry_doi:
            # Crossref API is designed for programmatic access
            resp = requests.get(
                f"https://api.crossref.org/works/{entry_doi}",
                timeout=10,
            )
            status = "OK" if resp.ok else f"ERROR {resp.status_code}"
            print(
                f"  [{bib_entry['ENTRYTYPE']}] {bib_entry['ID']}:"
                f" https://doi.org/{entry_doi} -> {status}"
            )
        else:
            resp = requests.get(
                url,
                headers={"User-Agent": "Mozilla/5.0"},
                allow_redirects=True,
                timeout=10,
            )
            status = "OK" if resp.ok else f"ERROR {resp.status_code}"
            print(
                f"  [{bib_entry['ENTRYTYPE']}] {bib_entry['ID']}:"
                f" {url} -> {status}"
            )
    except requests.RequestException as exc:
        print(
            f"  [{bib_entry['ENTRYTYPE']}] {bib_entry['ID']}: ERROR {exc}"
        )


with open("refs.bib", encoding="utf-8") as f:
    library = bibtexparser.load(f)

print(f"Entries in refs.bib ({len(library.entries)} total):")
for entry in library.entries:
    validate_bib_entry(entry)
`````

## Generating and typesetting tables

Tables in Quarto can be written directly in **Markdown**, generated from **Python**, or typeset using **LaTeX** and **HTML/CSS** for finer control. A useful distinction is between **generating** a table (creating it from data) and **typesetting** a table (controlling how it is displayed). In practice, data should usually be generated automatically from code and then formatted appropriately for the target output.

For simple tables, Markdown syntax is often sufficient. Markdown tables also support simple column alignment by placing colons in the separator row:

```markdown
| Quantity | Value | Unit |
|:---|---:|:---:| <!-- separator row - do alignment here -->
| Mass | 1.23 | [kg] |
| Time | 0.015 | [s] |
| Energy | 1.23e+03 | [J] |
```

> **Tip**
Here, `:---` left-aligns a column, `---:` right-aligns it, and `:---:` centres it. Right alignment is usually preferable for numerical columns, although Markdown cannot align numbers at the decimal separator. Markdown tables are easy to write and portable across formats, but offer limited control over alignment, text wrapping, and numerical formatting.

For computed results, tables should generally be generated from Python (e.g. with `pandas`), ensuring that the displayed values remain consistent with the underlying calculations:

`````python
# |echo: true
import pandas as pd

df = pd.DataFrame(
    {
        "Mass / [kg]": [1.234, 12.345, 1234.5],
        "Time / [s]": [0.01523, 1.2, 3600],
        "Energy / [J]": [12.3456, 1234.56, 1.23e6],
    }
)
`````

When rendered in Quarto or a Jupyter notebook, the `DataFrame` is displayed automatically as a table:

|    | Mass / [kg] | Time / [s] | Energy / [J] |
| -: | --------: | -------: | ---------: |
|  0 |     1.234 |  0.01523 |    12.3456 |
|  1 |    12.345 |      1.2 |    1234.56 |
|  2 |    1234.5 |     3600 |   1.23e+06 |

This default representation is useful for checking calculations, but it is not necessarily well suited for communication. The numerical precision is inconsistent, scientific notation follows programming conventions, and values are not formatted according to scientific or typographic conventions.

For publication-quality output, the table can be typeset explicitly:

```latex
\begin{tabular}{S S S}
\toprule
{Mass / kg} & {Time / s} & {Energy / J} \\
\midrule
1.23 & 0.0152 & 12.3 \\
12.3 & 1.20 & 1.23 \times 10^{3} \\
1.23 \times 10^{3} & 3.60 \times 10^{3} & 1.23 \times 10^{6} \\
\bottomrule
\end{tabular}
```

The formatted version is easier to read because numerical values are rounded consistently, scientific notation is displayed using mathematical notation, and the columns can be aligned for easier comparison.

Several formatting choices improve readability:

* Align numerical columns consistently, ideally at the decimal separator.
* Round values to an appropriate number of **significant digits** rather than displaying all available digits.
* Format scientific notation, e.g. convert `1.23e+03` into $1.23\times10^3$.
* Display time values in a suitable form, such as `15.2 ms`, `1.20 s`, or `01:00:00`.
* Separate quantities from units by placing units in the column header rather than repeating them in each cell.

For example, decimal alignment makes it easier to compare values:

```text
Not aligned:
1.2      [kg]
12.34    [kg]
123.456  [kg]

Decimal aligned:
  1.2    [kg]
 12.34   [kg]
123.456  [kg]
```

For PDF output, the LaTeX package `siunitx` is particularly useful because it can automatically align numbers, round values, control decimal markers, and format scientific notation:

```latex
\num{1.23e3}
```

which renders as $1.23\times10^3$.

Headers deserve special attention. They should identify both the quantity and its unit, for example `Mass / [kg]`, `Length / [m]`, or `Velocity / `[$m/s^{-1}$]. This keeps the table compact and avoids repeating units in every row.

Complex headers can be created by merging or splitting cells. In LaTeX this is typically done with `\multicolumn` and the `multirow` package:

```latex
\multicolumn{2}{c}{Measurement}
```

Long text inside a cell may require wrapping across multiple lines. Common solutions include fixed-width columns (`p{...}`), `tabularx`, or the `makecell` package. For example:

```latex
\begin{tabular}{p{3cm} S}
```

creates a fixed-width text column that wraps automatically.

> **Tip**
Formatting requirements may also depend on the output medium. Tables intended for reports can contain more detail and precise numerical alignment, whereas tables intended for slides should usually contain fewer columns, larger text, and reduced numerical precision.

Finally, Quarto supports additional styling through HTML and CSS. While LaTeX generally provides the highest-quality PDF tables, HTML styling can be useful for web pages and slides, allowing custom colours, fonts, borders, row highlighting, responsive layouts, and interactive elements.

In summary, use Markdown for small manually written tables, Python for reproducible data generation, LaTeX (especially `siunitx`) for precise numerical typesetting in reports, and HTML/CSS when richer web-based styling is required.

Vector graphics are the preferred format for plots, diagrams, maps, charts, and most visualisations
  in technical and professional documents.
Unlike raster graphics (e.g., PNG or JPG), which are composed of pixels, vector graphics (e.g., SVG or PDF) describe
  shapes mathematically and represent text as structured elements.
As a result, they are resolution-independent and can be scaled without loss of quality.

Screenshots and raster exports (e.g., PNG or JPG) should be avoided for figures whenever
  a vector format is available.
They should only be used for snapshots of software interfaces or content that cannot reasonably
  be represented in vector form.
Raster formats are primarily suitable for photographic images or other inherently pixel-based data.
Even for raster images, all text (labels, titles, annotations) should be added as vector graphics
  by embedding the images in SVG or PDF files which may contain a mixture of both vector and raster content.
Raster exports of plots, maps and diagrams introduce loss of scalability, non-selectable text,
  and reduced clarity when printed or zoomed.
  
Vector graphics are the expected format in academic journals and a best practice for professional materials
  across industries, including logo design and visual identity systems.
SVG is a widely supported vector format viewable in all modern web browsers and editable using open-source
  tools such as Inkscape, making it highly portable and interoperable.

### **Sankey Diagrams**
Sankey diagrams visualize the flow by defining source and target connections. We can use it for displaying flow of data in our database apps, flow of resources in a company, or any other flow of things. In Plotly, the structure is divided into two main components: **Nodes** and **Links**.[^plotly-sankey]

**Node Attributes**
Nodes represent the entities or stages (the vertical blocks) in your diagram.

> * **`label`**: List of Strings. Defines the name of each node. The position in this list determines the node's index (starting from 0).
> * **`thickness`**: Integer. Sets the width of the node rectangle in pixels.
> * **`pad`**: Integer. Sets the vertical padding between nodes in the same column.
> * **`color`**: String or List. The color of the nodes. You can provide one color for all or a list to color each node individually.
> * **`line`**: Dictionary. Defines the border of the node.
> * **`example`**: `line = dict(color = "black", width = 0.5)`

**Link Attributes**
Links represent the "ribbons" or flows between the nodes.

> * **`source`**: List of Integers. The index of the node where the flow starts.
> * **`target`**: List of Integers. The index of the node where the flow ends.
> * **`value`**: List of Numbers. Sets the volume/quantity of the flow. This determines the thickness of the ribbon.
> * **`color`**: String or List. The color of the flow ribbons (often set to be semi-transparent).
> * **`hoverlabel`**: Dictionary. Customizes the tooltip appearance when hovering over a flow.

**Layout Attributes**
Used within `fig.update_layout()` to control the canvas.

> * **`title_text`**: String. The title of the diagram.
> * **`font_size`**: Integer. The global font size for labels.
> * **`orientation`**: 'h' or 'v'. Choose between a horizontal flow (default) or a vertical flow.

[^plotly-sankey]: Plotly Development Team (2026). *Sankey diagram*. Plotly Graphing Libraries. Available at: <https://plotly.com/python/sankey-diagram/> (Accessed: June 20, 2026).

`````python
# | echo: false
# | label: fig-sankey-1
# | fig-cap: "Sankey diagram"
import plotly.graph_objects as go
from IPython.display import SVG, display

fig = go.Figure(
    data=[
        go.Sankey(
            node={
                "thickness": 10,
                "line": {"color": "green", "width": 0.1},
                "label": ["A", "B", "C", "D", "E", "F"],
                "color": [
                    "blue",
                    "blue",
                    "red",
                    "green",
                    "green",
                    "cyan",
                ],
            },
            link={
                # indices correspond to labels so:
                "source": [0, 5, 1, 4, 2, 3, 4],
                "target": [2, 1, 5, 2, 1, 5, 3],
                "value": [7, 1, 3, 6, 9, 4, 2],
                # A -> C: 7,
                # F -> B: 1,
                # B -> F: 3,
                # E -> C: 6,
                # C -> B: 9,
                # D -> F: 4,
                # E -> D: 2
                "color": [
                    "rgba(165, 42, 42, 0.5)",  # brown
                    "rgba(255, 192, 203, 0.5)",  # pink
                    "rgba(128, 128, 128, 0.5)",  # grey
                    "rgba(255, 192, 203, 0.5)",  # pink
                    "rgba(0, 128, 0, 0.5)",  # green
                    "rgba(255, 0, 0, 0.5)",  # red
                    "rgba(0, 0, 255, 0.5)",  # blue
                ],
            },
        )
    ]
)
fig.update_layout(
    title_text="Basic Sankey Diagram",
    font_size=14,  # Slightly larger font for readability
    width=600,  # Explicit pixel width for the slide
    height=300,  # Explicit pixel height for the slide
    margin={
        "l": 20,
        "r": 20,
        "t": 60,
        "b": 20,
    },
    paper_bgcolor='white',
)
fig.write_image("sankey1.svg")
display(SVG("sankey1.svg"))
`````

### **Ishikawa (Fishbone) Diagram**
The Ishikawa diagram is a visualization tool used to identify, explore, and display the potential causes of a specific problem or quality defect. It organizes thoughts into a "fishbone" structure to find the root cause of an issue.

---

**Core Structure**
The diagram is built around four main elements:

> * **The Head**: Represents the **Effect** or the problem statement being analyzed.
> * **The Backbone**: The central line connecting the causes to the effect.
> * **The Ribs (Main Branches)**: Represent the primary categories of causes.
> * **The Bones (Sub-branches)**: Represent the specific individual causes within those categories.

---

**The 6M Framework (Standard Categories)**
In manufacturing and business, the "ribs" of the diagram usually follow the **6Ms**:

> * **`Manpower`**: Personnel, training, and human factors.
> * **`Methods`**: Processes, rules, or production steps.
> * **`Machines`**: Tools, software, and equipment reliability.
> * **`Materials`**: Raw materials, components, or information quality.
> * **`Measurements`**: Data collection, standards, and inspection accuracy.
> * **`Mother Nature`**: Environment, temperature, and workspace conditions.

---

Unfortunately Ishikawa diagram doesn't have one simple implementation ready to use after importing a package. The solution below is a little complicated, but it gives us ready for use Ishikawa diagram.[^matplotlib-ishikawa]

[^matplotlib-ishikawa]: Matplotlib Development Team (2026). *Ishikawa diagram*. Matplotlib Stable Gallery. Available at: <https://matplotlib.org/stable/gallery/specialty_plots/ishikawa_diagram.html> (Accessed: May 11, 2026).

`````python
# | include: false
import math

from matplotlib.patches import Polygon, Wedge


def problems(
    data: str,
    problem_x: float,
    problem_y: float,
    angle_x: float,
    angle_y: float,
):
    """
    Draw each problem section of the Ishikawa plot.

    Parameters
    ----------
    data : str
        The name of the problem category.
    problem_x, problem_y : float, optional
        The `X` and `Y` positions of the problem arrows (`Y` defaults to zero).
    angle_x, angle_y : float, optional
        The angle of the problem annotations. They are always angled towards
        the tail of the plot.

    Returns
    -------
    None.

    """
    ax.annotate(
        str.upper(data),
        xy=(problem_x, problem_y),
        xytext=(angle_x, angle_y),
        fontsize=10,
        color='white',
        weight='bold',
        xycoords='data',
        verticalalignment='center',
        horizontalalignment='center',
        textcoords='offset fontsize',
        arrowprops={"arrowstyle": "->", "facecolor": 'black'},
        bbox={
            "boxstyle": 'square',
            "facecolor": 'tab:red',
            "pad": 0.8,
        },
    )


def causes(
    data: list,
    cause_x: float,
    cause_y: float,
    cause_xytext=(-9, -0.3),
    top: bool = True,
):
    """
    Place each cause to a position relative to the problems
    annotations.

    Parameters
    ----------
    data : indexable object
        The input data. IndexError is
        raised if more than six arguments are passed.
    cause_x, cause_y : float
        The `X` and `Y` position of the cause annotations.
    cause_xytext : tuple, optional
        Adjust to set the distance of the cause text from the problem
        arrow in fontsize units.
    top : bool, default: True
        Determines whether the next cause annotation will be
        plotted above or below the previous one.

    Returns
    -------
    None.

    """
    for index, cause in enumerate(data):
        # [<x pos>, <y pos>]
        coords = [
            [0.02, 0],
            [0.23, 0.5],
            [-0.46, -1],
            [0.69, 1.5],
            [-0.92, -2],
            [1.15, 2.5],
        ]

        # First 'cause' annotation is placed in the middle of the 'problems' arrow
        # and each subsequent cause is plotted above or below it in succession.
        cause_x -= coords[index][0]
        cause_y += coords[index][1] if top else -coords[index][1]

        ax.annotate(
            cause,
            xy=(cause_x, cause_y),
            horizontalalignment='center',
            xytext=cause_xytext,
            fontsize=9,
            xycoords='data',
            textcoords='offset fontsize',
            arrowprops={"arrowstyle": "->", "facecolor": 'black'},
        )


def draw_body(data: dict):
    """
    Place each problem section in its correct place by changing
    the coordinates on each loop.

    Parameters
    ----------
    data : dict
        The input data (can be a dict of lists or tuples). ValueError
        is raised if more than six arguments are passed.

    Returns
    -------
    None.

    """
    # Set the length of the spine according to the number of 'problem' categories.
    length = (math.ceil(len(data) / 2)) - 1
    draw_spine(-2 - length, 2 + length)

    # Change the coordinates of the 'problem' annotations after each one is rendered.
    offset = 0
    prob_section = [1.55, 0.8]
    for index, problem in enumerate(data.values()):
        plot_above = index % 2 == 0
        cause_arrow_y = 1.7 if plot_above else -1.7
        y_prob_angle = 16 if plot_above else -16

        # Plot each section in pairs along the main spine.
        prob_arrow_x = prob_section[0] + length + offset
        cause_arrow_x = prob_section[1] + length + offset
        if not plot_above:
            offset -= 2.5
        if index > 5:
            raise ValueError(
                f'Maximum number of problems is 6, you have entered '
                f'{len(data)}'
            )

        problems(
            list(data.keys())[index],
            prob_arrow_x,
            0,
            -12,
            y_prob_angle,
        )
        causes(problem, cause_arrow_x, cause_arrow_y, top=plot_above)


def draw_spine(xmin: int, xmax: int):
    """
    Draw main spine, head and tail.

    Parameters
    ----------
    xmin : int
        The default position of the head of the spine's
        x-coordinate.
    xmax : int
        The default position of the tail of the spine's
        x-coordinate.

    Returns
    -------
    None.

    """
    ax.plot([xmin - 0.1, xmax], [0, 0], color='tab:red', linewidth=2)
    ax.text(
        xmax + 0.1,
        -0.05,
        'PROBLEM',
        fontsize=10,
        weight='bold',
        color='white',
    )
    semicircle = Wedge((xmax, 0), 1, 270, 90, fc='tab:red')
    ax.add_patch(semicircle)
    tail_pos = [[xmin - 0.8, 0.8], [xmin - 0.8, -0.8], [xmin, -0.01]]
    triangle = Polygon(tail_pos, fc='tab:red')
    ax.add_patch(triangle)


# Input data
`````

### **Radar Chart (Spider Chart)**
Radar charts are used to compare multiple quantitative variables across different categories. They are particularly effective for showing performance profiles, strengths, and weaknesses relative to a central point.[^plotly-radar]

---

**Core Attributes (go.Scatterpolar)**
In Plotly, radar charts are created using the `Scatterpolar` object.

> * **`r`**: *List of Numbers*. Represents the **radial** distance (values) from the center for each category. Ensure it's repeated at the end of the list.
> * **`theta`**: *List of Strings*. Represents the **angular** categories (labels) around the chart. Ensure it's repeated at the end of the list.
> * **`fill`**: *'toself' or 'none'*. Determines if the area inside the chart lines should be filled with color.
> * **`name`**: *String*. Sets the trace name, which appears in the legend when comparing multiple datasets.
> * **`line`**: *Dictionary*. Controls the color and width of the connecting lines (e.g., `line=dict(color='blue')`).

**Layout Attributes (polar)**
To properly view a radar chart, the layout must include a `polar` configuration.

> * **`radialaxis`**: Controls the scale of the values (e.g., `visible=True`, `range=[0, 100]`).
> * **`angularaxis`**: Controls the appearance of the outer labels and category lines.
> * **`gridshape`**: *'circular' or 'linear'*. Defines whether the background grid looks like a circle or a polygon (spider web).

[^plotly-radar]: Plotly Development Team (2026). *Radar chart*. Plotly Graphing Libraries. Available at: <https://plotly.com/python/radar-chart/> (Accessed: June 20, 2026).

`````python
# | echo: false
# | label: fig-radar-2
# | fig-cap: "Radar diagram"
categories = [
    'processing cost',
    'mechanical properties',
    'chemical stability',
    'thermal stability',
    'device integration',
]
fig = go.Figure()
# To add additional traces, we need to use go.Figure().add_trace like shown below:
fig.add_trace(
    go.Scatterpolar(
        r=[1, 5, 2, 2, 3],
        theta=categories,
        fill='toself',
        name='Product A',
    )
)
fig.add_trace(
    go.Scatterpolar(
        r=[4, 3, 2.5, 1, 2],
        theta=categories,
        fill='toself',
        name='Product B',
    )
)
fig.update_layout(
    width=600,
    height=400,
    font_size=14,
    polar={
        "radialaxis": {"visible": True, "range": [0, 5]},
        "angularaxis": {
            "rotation": 50,
            "linecolor": 'red',  # Border color
            "gridcolor": 'red',  # Color of lines linking the center of radar with it's border
        },
        "gridshape": 'linear',
    },
    showlegend=False,
    paper_bgcolor='white',
    plot_bgcolor='white',
    font_color="black",
)
fig.write_image("radar2.svg")
display(SVG("radar2.svg"))
`````

## Visualising networks with \techword{NetworkX}


\techword{NetworkX} is a Python library for creating, manipulating,
  and studying graphs - mathematical structures that model pairwise relationships.
A graph consists of **nodes** (vertices) connected by **edges** (links).
Edges may be **undirected** (symmetric) or **directed** (one-way, drawn as arrows),
  and may carry numeric **weights** encoding cost, distance, or strength.
A **path** is a sequence of edges connecting two nodes;
  classic algorithms such as Dijkstra's shortest path (`nx.shortest_path`)
  or PageRank (`nx.pagerank`) operate on these structures.

`````python
# | echo: true
# | code-fold: true
# | code-summary: "Graph creation code"

import networkx as nx

rng = np.random.default_rng(1)
G = nx.gnm_random_graph(8, 18, seed=1, directed=True)
G = nx.relabel_nodes(G, {i: chr(65 + i) for i in range(8)})
nx.set_edge_attributes(
    G,
    {e: int(rng.integers(1, 6)) for e in G.edges()},
    "weight",
)
`````

@fig-pagerank illustrates a randomly generated weighted digraph
  (`{python} G.number_of_nodes()` nodes, `{python} G.number_of_edges()` edges)
  where each node's diameter and fill colour
  encode its PageRank centrality score -
  a metric originally developed for ranking web pages [@Brin_Page_1998].
Edge thickness is proportional to the weight attribute,
  and a `ScalarMappable` colorbar maps PageRank values
  to the `YlOrRd` colourmap.
All visual defaults - node size (default 300), edge width (1.0),
  font size, colours - are controlled via keyword arguments
  to `draw_networkx_nodes`, `draw_networkx_edges`,
  and related functions;
  refer to the \techword{NetworkX} drawing-module documentation
  for the full parameter list.

`````python
# | echo: true
# | code-fold: true
# | code-summary: "PageRank visualisation code"
# | label: fig-pagerank
# | fig-cap: "Weighted directed graph coloured by PageRank score."

pr = nx.pagerank(G, weight="weight")
pos = nx.spring_layout(G, seed=21, k=1.8)
vals = np.array([pr[n] for n in G])
cmap, norm = pyplot.cm.YlOrRd, pyplot.Normalize(
    vals.min(), vals.max()
)
wt = nx.get_edge_attributes(G, "weight")

fig, ax = pyplot.subplots()
nx.draw_networkx_edges(
    G,
    pos,
    ax=ax,
    width=[wt[e] * 0.6 for e in G.edges()],
    edge_color="#aaa",
    arrowstyle="-|>",
    arrowsize=18,
    connectionstyle="arc3,rad=0.1",
    alpha=0.7,
    min_source_margin=18,
    min_target_margin=18,
)
nx.draw_networkx_nodes(
    G,
    pos,
    ax=ax,
    node_color=[cmap(norm(pr[n])) for n in G],
    node_size=400
    + (vals - vals.min()) / (vals.max() - vals.min()) * 1800,
    edgecolors="w",
    linewidths=1.2,
)
nx.draw_networkx_labels(
    G, pos, ax=ax, font_size=11, font_weight="bold"
)
pyplot.colorbar(
    pyplot.cm.ScalarMappable(cmap=cmap, norm=norm),
    ax=ax,
    fraction=0.03,
    pad=0.02,
    label="PageRank score",
)
ax.axis("off");
`````

## Creating and annotating maps with \techword{Folium} and \techword{OSMnx}

Maps produced programmatically from OpenStreetMap data combine the flexibility of
\techword{Python} with freely available, community-maintained geodata.
The example below demonstrates geographic projections, layers, interactivity,
colour customisation, OSM licensing, and offline caching -- using the
\href{https://www.agh.edu.pl}{AGH University} campus in Krak\'{o}w as a real-world case.

**Geographic projections.**
All spatial data carries a *coordinate reference system* (CRS).
OpenStreetMap stores coordinates in WGS~84 (EPSG:4326), the global latitude/longitude
system used by GPS.
Web tile providers render in Web Mercator (EPSG:3857), which preserves angles locally
but distorts area at high latitudes.
\techword{OSMnx} and \techword{GeoPandas} handle CRS reprojection transparently via \techword{PROJ}.

**Layers.**
\techword{Folium} wraps the JavaScript \techword{Leaflet} library and supports multiple
independently toggleable layers: raster tile basemaps, GeoJSON polygons, and point
markers can all coexist in one map and be switched on or off by the reader.
For static PDF output, \techword{Contextily} combined with \techword{Matplotlib}
composites tile images directly onto a Matplotlib axes object.

**Interactivity.**
HTML maps generated by \techword{Folium} support panning, zooming,
click-triggered popups, and hover tooltips without any server-side infrastructure --
well suited for digital appendices and online supplementary materials.

**Colour choices.**
Choropleth maps benefit from perceptually uniform, colour-blind-safe palettes such as
\techword{ColorBrewer} or \techword{Matplotlib}'s \texttt{viridis}/\texttt{cividis}.
Sequential palettes suit single-variable gradients; diverging palettes suit data with
a meaningful midpoint.

**Licensing and attribution.**
OpenStreetMap data is released under the Open Database License (ODbL~1.0):
derivative works must carry attribution and remain open if redistributed~[@OpenStreetMap].
Tile images are CC~BY-SA~2.0.

## Embedding musical notation with \techword{MusiX\TeX}

Scores can be typeset from the thesis source via \techword{MusiX\TeX} and its \techword{PMX} preprocessor. The direct route - `\usepackage{musixtex}` with raw \LaTeX{} - tends to clash with \techword{Quarto}'s template, so we drive the external toolchain (\techword{pmxab}~\textrightarrow~\techword{musixtex}~\textrightarrow~\techword{pdfcrop}) from a&nbsp;notebook cell and embed the cropped \techword{PDF}. Rather than running `pmxab`, `musixtex`, and `pdfcrop` by hand every time a note is tweaked, the helper `render_pmx(source, caption, width)` chains all three calls in one place; it is defined once in a hidden setup cell, and calling it again with a fresh name is all that is required to add a new score. See examples below, and review the source code for full insight!

`````python
# | include: false
import os
import shutil
import subprocess
from hashlib import sha1
from pathlib import Path
from tempfile import TemporaryDirectory

from IPython.display import Markdown

# put TinyTeX's binaries on PATH (pmxab, musixtex, pdfcrop ship with the standard distribution)
for candidate in (
    Path.home() / "Library/TinyTeX/bin/universal-darwin",  # macOS
    Path.home() / ".TinyTeX/bin/x86_64-linux",  # Linux x86_64
    Path.home() / ".TinyTeX/bin/aarch64-linux",  # Linux ARM
    Path.home() / "AppData/Roaming/TinyTeX/bin/windows",  # Windows
):
    if candidate.is_dir():
        os.environ["PATH"] = (
            f"{candidate}{os.pathsep}{os.environ['PATH']}"
        )
        break


def render_pmx(*, source: str, caption: str, width: str) -> Markdown:
    """Compile PMX source to a cropped PDF and return a Markdown figure.

    The PDF filename is derived from the PMX source, so identical inputs
    reuse the same file. Compilation is performed in a temporary directory;
    only the final cropped PDF is written to the current directory.
    """
    stem = f"pmx-{sha1(source.encode('utf-8')).hexdigest()[:12]}"
    out_pdf = Path(f"{stem}.pdf")

    # Skip compilation if we've already generated this score.
    if not out_pdf.exists():
        with TemporaryDirectory() as tmpdir:
            tmp = Path(tmpdir)

            (tmp / f"{stem}.pmx").write_text(source, encoding="utf-8")
            run_args = {
                'cwd': tmp,
                'capture_output': True,
            }
            subprocess.run(["pmxab", stem], check=True, **run_args)
            subprocess.run(["musixtex", stem], check=True, **run_args)
            subprocess.run(
                ["pdfcrop", f"{stem}.pdf", f"{stem}.pdf"],
                check=True,
                **run_args,
            )

            shutil.copy2(tmp / f"{stem}.pdf", out_pdf)

    return Markdown(f"![{caption}]({out_pdf}){{width={width}}}")
`````

`````python
# | include: true
render_pmx(
    width="70%",
    caption="One instrument, two score lines",
    source=r"""
1 1 4 4 4 4 0 0
1 2 16 0

t
./
c4 d e f /
g a b c /   
c b a g /
f e d c /    
/
""",
)
`````

`````python
# | include: true

render_pmx(
    caption="Two instruments, treble + treble, D major",
    width="55%",
    # After the header, one blank line per instrument - two here.
    # Clef string `tt` lists staves bottom-up; line ends at first space.
    source=r"""
2 2 3 4 3 4 0 2
1 1 16 0


tt
./
c4 d e /
c45 d e /
g4 a g /
g45 a g /
/
""",
)
`````

`````python
render_pmx(
    caption="Piano grand staff (one instrument, two staves)",
    width="55%",
    # Clef string `bt` = bass on the lower staff, treble on the upper
    # One blank line - both staves belong to the same instrument.
    source=r"""
2 1 4 4 4 4 0 0
1 1 16 0

bt
./
c43 d e f /
c45 d e f /
g4 a g f /
g4 a g f /
/
""",
)
`````

**Below you can find an explanation of the \techword{PMX} header numbers.**

```
1 1 4 4 4 4 0 0 ← line 1
1  - number of voices
1  - number of instruments
4  - logical meter numerator (beats per measure)
4  - logical meter denominator (note value of the beat)
4  - printed meter numerator
4  - printed meter denominator
0  - pickup-bar length, in beats (0 = no pickup)
0  - key signature (+n = n sharps, -n = n flats, 0 = C major / A minor)

1 1 20 0 ← line 2
1  - number of pages (0 = let PMX decide)
1  - systems per piece (if npages>0); else average bars per system
20 - staff size in points (16, 20, 24, or 29; 20 is the default)
0  - indent of the first system, as a fraction of the line width (0 = none)
t  - active clef (treble); one letter per staff, listed bottom-up
./ - start of the note stream
```

A few things to keep in mind:

- *Note names.* \techword{PMX} uses `a`--`g` (Anglo-Saxon), `b` = B-natural. The Polish `H`/`B` convention is **not** supported -- transcribe `H` $\to$ `b`, Polish `B` $\to$ `bf`.
- *Multi-instrument.* Raise `voices`/`instruments` on line 1 and add one blank line per instrument before the clef string; for a piano grand staff use `instruments < voices`.
- *Common pitfalls.* Missing trailing `/` at bar ends; \techword{PMX} comments use `%` in column 1, not `#`; leftover `*.mx1`/`*.mx2` files should sit in `.gitignore`.

**Further reading.** The official manuals are the authoritative reference but tend to be&nbsp;terse; the community resources below are often a friendlier starting point:

- [\techword{PMX} package on CTAN](https://ctan.org/pkg/pmx) -- includes the full PMX user guide PDF by Don Simons.
- [\techword{MusiX\TeX} package on CTAN](https://ctan.org/pkg/musixtex) -- the underlying typesetter; consult its manual for \techword{\TeX}-level customisation (`\setname`, custom clefs, etc.).
- [`tex.stackexchange.com` -- musixtex tag](https://tex.stackexchange.com/questions/tagged/musixtex) -- real-world Q&A; usually the fastest way to&nbsp;debug an unfamiliar PMX error.

## Making your code publicly available as a package at \techword{PyPI.org}

### Project Configuration (`pyproject.toml`)
Place a `pyproject.toml` file in the root directory of your project. This file contains the metadata about your package and specifies the build system. Here is a minimal example:

```toml
[build-system]
requires = ["setuptools>=61.0"]
build-backend = "setuptools.build_meta"

[project]
name = "my_thesis"
version = "0.1.0"
authors = [
  { name="Your Name", email="your.email@student.agh.edu.pl" },
]
description = "Code repository for my Master's thesis"
readme = "README.md"
requires-python = ">=3.9"
dependencies = [
    "numpy",
    "matplotlib",
]
```
- The `[build-system]` section acts as an instruction manual for the packaging tool. 
  - `requires`: Informs the system to ensure the required build tools are installed before attempting to build the package.
  - `build-backend`: Specifies the exact engine within setuptools that will physically handle the creation of the distribution archives.
- The `[project]` section acts as the identity card for your project.
  - `name`: The identifier. This name must be absolutely unique across the entire PyPI registry.
  - `version`: The current version of your code.
  - `authors`: A list of creators.
  - `description`: A concise summary of the package, visible in PyPI search results.
  - `readme`: Instructs PyPI to parse the specified file and display it as the description on the package's main page.
  - `requires-python`: Prevents installation on incompatible, older Python environments.
  - `dependencies`: Listing external libraries ensures that when someone installs your package, pip automatically fetches and installs these prerequisites first

### Building the Distribution
Ensure you have the build package installed (`pip install build`), and run the following command from the same directory where `pyproject.toml` is located:
```bash
python -m build
```
This will create a `dist/` directory containing two critical files: a Source Distribution (`.tar.gz`) and a built Wheel (`.whl`), which is an optimised format that allows users to install your package in a fraction of a second without compiling it locally.

### Uploading to PyPI
To upload your package safely, use twine (`pip install twine`). It is recommended to test the process using a testing repository before publishing to the main repository.

```bash
# Uploading to the testing repository
python -m twine upload --repository testpypi dist/*

# Uploading to the real repository
python -m twine upload dist/*
```
*   **Authentication Note:** For security reasons, \techword{PyPI} no longer accepts standard passwords. When prompted by twine, you must use `__token__` as your username and provide your generated API Token as your password.

## Version numbering and using \techword{git} metadata for package versions


Version numbers play an important role in software development because they allow developers and users to&nbsp;identify the exact revision of a package. They make it easier to&nbsp;handle debugging, dependency management, release management, and reproducibility. In research projects, version information makes it possible to determine which software revision was used to produce a particular result.

A commonly used versioning scheme is **Semantic Versioning (*SemVer*)**, which follows the format `MAJOR.MINOR.PATCH`. The MAJOR number increases when changes break backward compatibility, `MINOR` when new *backward-compatible functionality* is added, and&nbsp;`PATCH` when *backward-compatible bug fixes* are released. This convention allows users to&nbsp;estimate the impact of an update by examining the version number.

Version numbers can be maintained *manually*, for example by storing a value such as&nbsp;`version = "1.4.2"` in project metadata. However, this approach is prone to&nbsp;*human error* because developers often forget to update the version before a release, create inconsistent version numbers across branches, or introduce merge conflicts. As&nbsp;projects grow and release more frequently, ***manual version management* becomes *increasingly difficult***. To avoid these problems, package versions **can be generated automatically** from *\techword{Git} metadata*. In this approach, release versions are associated with *\techword{Git} tags*, while development versions may additionally include information derived from *commits*. For&nbsp;example, a repository tagged as `v1.2.0` will generate version `1.2.0`, where a&nbsp;build created several commits later may have version such as `1.2.1.dev5+gabc123`. Such version names make it possible to trace a package directly to the *corresponding repository state*.

In \techword{Python} projects, this functionality is provided by \techword{setuptools-scm}. One of the&nbsp;main advantages is the simplicity of its configuration. Instead of specifying a fixed version number, the project declares that the version should be generated dynamically:

```{=latex}
\Needspace{14\baselineskip}
\noindent\techword{pyproject.toml}\par\vspace{0.35\baselineskip}
\begin{Shaded}
\begin{Verbatim}[fontsize=\small]
[build-system]
requires = ["setuptools>=80", "setuptools-scm[simple]>=9.2"]
build-backend = "setuptools.build_meta"

[project]
name = "example"
dynamic = ["version"]
\end{Verbatim}
\end{Shaded}
```

The generated version must still comply with \techword{Python} packaging standards **(PEP 440)**. Raw \techword{Git} commit hashes cannot serve as the main version number when publishing to&nbsp;\techword{PyPI}; they may appear only as local identifiers after `+` in development builds (as&nbsp;in&nbsp;`1.2.1.dev5+gabc123`). A&nbsp;consistent tagging strategy is then necessary to ensure meaningful and valid version numbers.

***Alternative tools*** for \techword{Git}-based versioning include \techword{versioneer}, \techword{hatch-vcs}, and&nbsp;\techword{poetry-dynamic-versioning}. However, \techword{setuptools-scm} is widely used due to&nbsp;its&nbsp;minimal configuration requirements and seamless integration with standard \techword{Python} packaging workflows.


## Making your docs available on \techword{GitHub Pages} or \techword{Read The Docs}


### Automation and Industry Standards
Manual documentation rendering is error-prone and leads to code-doc divergence. Automated build and deployment pipelines (\techword{CI/CD}) triggered on every repository commit represent the industry standard, ensuring that hosted documentation accurately reflects the active codebase. The two primary deployment platforms are:

* **\techword{GitHub Pages}** (native to \techword{GitHub}, requiring \techword{GitHub Actions} to deploy static \techword{HTML}).
* **\techword{Read the Docs}** (external hosting triggered via webhooks, requiring no \techword{GitHub Actions}). 

For complex projects, \techword{Read the Docs} provides multi-version support, allowing users to switch between stable releases (e.g., `v1.0`) and development branches (`latest`), which prevents integration errors.

### Tooling Trade-offs and Docstring Extraction

Using \techword{Quarto} establishes a single source of truth by embedding executable \techword{Python} code directly into \techword{Markdown}, but it lacks native mechanisms for deep, automated \techword{API} extraction from source code signatures. When documentation relies on docstrings (inline code comments), specialized generators eliminate manual formatting:

* **\techword{Sphinx}:** The industry standard for \techword{Python}; utilizes extensions like `autodoc` for deep \techword{API} reference generation.

* **\techword{MkDocs}:** A fast, \techword{Markdown}-centric generator configured via a single \techword{YAML} file.

* **\techword{pdoc}:** A zero-configuration tool that generates \techword{API} references directly from docstrings without configuration files.

### Workflow and Deployment Configurations
The following \techword{GitHub Actions} workflow (`.github/workflows/deploy-docs.yml`) automates rendering via \techword{Quarto} and deploys it to \techword{GitHub Pages}:

```yaml
name: Deploy Docs
on:
  push:
    branches: [main]
jobs:
  deploy:
    runs-on: ubuntu-latest
    permissions:
      pages: write
      id-token: write
    steps:
      - uses: actions/checkout@v4
      - run: pip install jupyter matplotlib pyyaml nbformat nbclient
      - uses: quarto-dev/quarto-actions/setup@v2
      - run: quarto render thesis.ipynb --to html
      - uses: actions/deploy-pages@v4
```

To configure \techword{Read the Docs} using `.readthedocs.yaml`, use the \techword{MkDocs} specification or implement custom execution jobs to run the \techword{Quarto} compiler directly:
