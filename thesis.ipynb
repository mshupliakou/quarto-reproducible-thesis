## Equations and mathematical typesetting guidelines

Quarto natively supports \techword{LaTeX} syntax for mathematical typesetting. Using proper math modes ensures that your equations are formatted clearly and cross-referenced correctly.

### Basic math modes

There are two primary ways to include mathematics in your document:

* **Inline equations:** Used within a paragraph. Enclose the expression in single dollar signs. 
    For example, typing `Energy is $E = mc^2$` renders as: Energy is $E = mc^2$.
* **Display equations:** Used for equations that require their own line. Enclose the expression in double dollar signs (`$$`). To automatically number the equation for cross-referencing, append a label `{#eq-label}` after the closing dollars.

```latex
$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot
\mathrm{e}^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$ {#eq-gauss}
```

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot
\mathrm{e}^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$ {#eq-gauss}

You can then reference it in your text by typing `@eq-gauss` (which renders as an active link: @eq-gauss).

### Quick reference: Common operators

Here is a quick cheat sheet for the most frequent mathematical structures:

| Element | Quarto (LaTeX) code | Rendered output |
| :--- | :--- | :--- |
| **Indexes** | `$x_i^2$` | $x_i^2$ |
| **Fractions** | `$\frac{a}{b}$` | $\frac{a}{b}$ |
| **Integrals** | `$\int_{0}^{\infty} x \,\mathrm{d}x$` | $\int_{0}^{\infty} x \,\mathrm{d}x$ |
| **Sums** | `$\sum_{n=1}^{N} n$` | $\sum_{n=1}^{N} n$ |

### Academic precision: The MIT guidelines

Writing math in a thesis is more than making it readable. Following the **MIT Mathematical Typesetting Guidelines** here are some basic tips:

* **Variables** (e.g., $x, y, m$) represent changing values and must always be **italicized** (LaTeX does this by default in math mode).
* **Constants, operators, and differentials** (e.g., Euler's number $\mathrm{e}$, differential $\mathrm{d}x$) are fixed and must be written in **upright (Roman) font**. We achieve this using the `\mathrm{}` command.
* **Standard functions** (e.g., $\sin, \exp, \ln$) should also be upright. Use built-in LaTeX commands with a backslash (`\sin`) instead of typing plain text. This applies both the correct upright font and proper spacing.

> **Tip: Fine-tuning - spacing:** Notice the `\,` command in the integrals example above. It adds a small space before the differential, which is highly recommended.

Moreover, it is important to remember that mathematical exposition consists of sentences. The usage of good grammar and punctuation does not change. For stylistic reasons avoid beginning sentences with symbols.

*For more typesetting guidelines, refer to the official [Mathematical Typesetting Guidelines](https://direct.mit.edu/DocumentLibrary/SubGuides/NECO-math-typesetting.pdf).*

*For advanced features, refer to the official [Overleaf Mathematical Expressions Guide](https://www.overleaf.com/learn/latex/Mathematical_expressions).*

### Typesetting with mhchem

The `mhchem` LaTeX package automatically applies proper IUPAC typesetting rules to plain text chemistry, avoiding the need for manual formatting.

* **Molecules:** `\ce{H2O}` translates to the standard upright water formula: $\ce{H2O}$
* **Ions:** `\ce{SO4^2-}` automatically formats the superscript charge: $\ce{SO4^2-}$
* **Isotopes:** `\ce{^{227}_{90}Th+}` aligns the mass and atomic numbers correctly: $\ce{^{227}_{90}Th+}$
* **Bonds:** `\ce{C=C}` and `\ce{N#N}` render as standard double and triple bonds: $\ce{C=C}$ and $\ce{N#N}$
* **Reactions:** `\ce{Zn^2+ <=>[- 2OH^-][- 2H^+] Zn(OH)2 v}` handles complex equilibrium arrows and states of matter: $\ce{Zn^2+ <=>[- 2OH^-][- 2H^+] Zn(OH)2 v}$

````python
# | label: tbl-stoichiometry-example
# | tbl-cap: "Automated calculations with Mendeleev and ChemPy"

# fetching precise, up-to-date elemental properties to avoid hardcoding
nitrogen = element('N')
print(f"Nitrogen atomic weight: {nitrogen.atomic_weight:.4f} g/mol")

# calculating molar mass from a formula automatically
water = Substance.from_formula('H2O')
print(f"Water molar mass: {water.mass:.3f} g/mol")

# balancing stoichiometry programmatically
reac, prod = balance_stoichiometry({'NH3', 'O2'}, {'NO', 'H2O'})
print(f"Balanced reactants: {dict(reac)}")
print(f"Balanced products:  {dict(prod)}")
````

## Generating and typesetting tables

Tables in Quarto can be written directly in **Markdown**, generated from **Python**, or typeset using **LaTeX** and **HTML/CSS** for finer control. A useful distinction is between **generating** a table (creating it from data) and **typesetting** a table (controlling how it is displayed). In practice, data should usually be generated automatically from code and then formatted appropriately for the target output.

For simple tables, Markdown syntax is often sufficient. Markdown tables also support simple column alignment by placing colons in the separator row:

```markdown
| Quantity | Value | Unit |
|:---|---:|:---:| <!-- separator row - do alignment here -->
| Mass | 1.23 | [kg] |
| Time | 0.015 | [s] |
| Energy | 1.23e+03 | [J] |
```

> **Tip**
Here, `:---` left-aligns a column, `---:` right-aligns it, and `:---:` centres it. Right alignment is usually preferable for numerical columns, although Markdown cannot align numbers at the decimal separator. Markdown tables are easy to write and portable across formats, but offer limited control over alignment, text wrapping, and numerical formatting.

For computed results, tables should generally be generated from Python (e.g. with `pandas`), ensuring that the displayed values remain consistent with the underlying calculations:

````python
# |echo: true
df = pd.DataFrame(
    {
        "Mass / [kg]": [1.234, 12.345, 1234.5],
        "Time / [s]": [0.01523, 1.2, 3600],
        "Energy / [J]": [12.3456, 1234.56, 1.23e6],
    }
)
````

When rendered in Quarto or a Jupyter notebook, the `DataFrame` is displayed automatically as a table:

|    | Mass / [kg] | Time / [s] | Energy / [J] |
| -: | --------: | -------: | ---------: |
|  0 |     1.234 |  0.01523 |    12.3456 |
|  1 |    12.345 |      1.2 |    1234.56 |
|  2 |    1234.5 |     3600 |   1.23e+06 |

This default representation is useful for checking calculations, but it is not necessarily well suited for communication. The numerical precision is inconsistent, scientific notation follows programming conventions, and values are not formatted according to scientific or typographic conventions.

For publication-quality output, the table can be typeset explicitly:

```latex
\begin{tabular}{S S S}
\toprule
{Mass / kg} & {Time / s} & {Energy / J} \\
\midrule
1.23 & 0.0152 & 12.3 \\
12.3 & 1.20 & 1.23 \times 10^{3} \\
1.23 \times 10^{3} & 3.60 \times 10^{3} & 1.23 \times 10^{6} \\
\bottomrule
\end{tabular}
```

The formatted version is easier to read because numerical values are rounded consistently, scientific notation is displayed using mathematical notation, and the columns can be aligned for easier comparison.

Several formatting choices improve readability:

* Align numerical columns consistently, ideally at the decimal separator.
* Round values to an appropriate number of **significant digits** rather than displaying all available digits.
* Format scientific notation, e.g. convert `1.23e+03` into $1.23\times10^3$.
* Display time values in a suitable form, such as `15.2 ms`, `1.20 s`, or `01:00:00`.
* Separate quantities from units by placing units in the column header rather than repeating them in each cell.

For example, decimal alignment makes it easier to compare values:

```text
Not aligned:
1.2      [kg]
12.34    [kg]
123.456  [kg]

Decimal aligned:
  1.2    [kg]
 12.34   [kg]
123.456  [kg]
```

For PDF output, the LaTeX package `siunitx` is particularly useful because it can automatically align numbers, round values, control decimal markers, and format scientific notation:

```latex
\num{1.23e3}
```

which renders as $1.23\times10^3$.

Headers deserve special attention. They should identify both the quantity and its unit, for example `Mass / [kg]`, `Length / [m]`, or `Velocity / `[$m/s^{-1}$]. This keeps the table compact and avoids repeating units in every row.

Complex headers can be created by merging or splitting cells. In LaTeX this is typically done with `\multicolumn` and the `multirow` package:

```latex
\multicolumn{2}{c}{Measurement}
```

Long text inside a cell may require wrapping across multiple lines. Common solutions include fixed-width columns (`p{...}`), `tabularx`, or the `makecell` package. For example:

```latex
\begin{tabular}{p{3cm} S}
```

creates a fixed-width text column that wraps automatically.

> **Tip**
Formatting requirements may also depend on the output medium. Tables intended for reports can contain more detail and precise numerical alignment, whereas tables intended for slides should usually contain fewer columns, larger text, and reduced numerical precision.

Finally, Quarto supports additional styling through HTML and CSS. While LaTeX generally provides the highest-quality PDF tables, HTML styling can be useful for web pages and slides, allowing custom colours, fonts, borders, row highlighting, responsive layouts, and interactive elements.

In summary, use Markdown for small manually written tables, Python for reproducible data generation, LaTeX (especially `siunitx`) for precise numerical typesetting in reports, and HTML/CSS when richer web-based styling is required.

Vector graphics are the preferred format for plots, diagrams, maps, charts, and most visualisations 
  in technical and professional documents.
Unlike raster graphics (e.g., PNG or JPG), which are composed of pixels, vector graphics (e.g., SVG or PDF) describe 
  shapes mathematically and represent text as structured elements.
As a result, they are resolution-independent and can be scaled without loss of quality.

Screenshots and raster exports (e.g., PNG or JPG) should be avoided for figures whenever 
  a vector format is available.
They should only be used for snapshots of software interfaces or content that cannot reasonably 
  be represented in vector form. 
Raster formats are primarily suitable for photographic images or other inherently pixel-based data.
Even for raster images, all text (labels, titles, annotations) should be added as vector graphics
  by embedding the images in SVG or PDF files which may contain a mixture of both vector and raster content. 
Raster exports of plots, maps and diagrams introduce loss of scalability, non-selectable text, 
  and reduced clarity when printed or zoomed.
  
Vector graphics are the expected format in academic journals and a best practice for professional materials 
  across industries, including logo design and visual identity systems.
SVG is a widely supported vector format viewable in all modern web browsers and editable using open-source 
  tools such as Inkscape, making it highly portable and interoperable.

## Embedding musical notation with \techword{MusiX\TeX}

Scores can be typeset from the thesis source via \techword{MusiX\TeX} and its \techword{PMX} preprocessor. The direct route - `\usepackage{musixtex}` with raw \LaTeX{} - tends to clash with \techword{Quarto}'s template, so we drive the external toolchain (\techword{pmxab}~\textrightarrow~\techword{musixtex}~\textrightarrow~\techword{pdfcrop}) from a&nbsp;notebook cell and embed the cropped \techword{PDF}. Rather than running `pmxab`, `musixtex`, and `pdfcrop` by hand every time a note is tweaked, the helper `render_pmx(source, caption, width)` chains all three calls in one place; it is defined once in a hidden setup cell, and calling it again with a fresh name is all that is required to add a new score. See examples below, and review the source code for full insight!

````python
# | include: false
import os
import shutil
import subprocess
from hashlib import sha1
from pathlib import Path
from tempfile import TemporaryDirectory

from IPython.display import Markdown

# put TinyTeX's binaries on PATH (pmxab, musixtex, pdfcrop ship with the standard distribution)
for candidate in (
    Path.home() / "Library/TinyTeX/bin/universal-darwin",  # macOS
    Path.home() / ".TinyTeX/bin/x86_64-linux",  # Linux x86_64
    Path.home() / ".TinyTeX/bin/aarch64-linux",  # Linux ARM
    Path.home() / "AppData/Roaming/TinyTeX/bin/windows",  # Windows
):
    if candidate.is_dir():
        os.environ["PATH"] = (
            f"{candidate}{os.pathsep}{os.environ['PATH']}"
        )
        break


def render_pmx(*, source: str, caption: str, width: str) -> Markdown:
    """Compile PMX source to a cropped PDF and return a Markdown figure.

    The PDF filename is derived from the PMX source, so identical inputs
    reuse the same file. Compilation is performed in a temporary directory;
    only the final cropped PDF is written to the current directory.
    """
    stem = f"pmx-{sha1(source.encode('utf-8')).hexdigest()[:12]}"
    out_pdf = Path(f"{stem}.pdf")

    # Skip compilation if we've already generated this score.
    if not out_pdf.exists():
        with TemporaryDirectory() as tmpdir:
            tmp = Path(tmpdir)

            (tmp / f"{stem}.pmx").write_text(source, encoding="utf-8")
            run_args = {
                'cwd': tmp,
                'capture_output': True,
            }
            subprocess.run(["pmxab", stem], check=True, **run_args)
            subprocess.run(["musixtex", stem], check=True, **run_args)
            subprocess.run(
                ["pdfcrop", f"{stem}.pdf", f"{stem}.pdf"],
                check=True,
                **run_args,
            )

            shutil.copy2(tmp / f"{stem}.pdf", out_pdf)

    return Markdown(f"![{caption}]({out_pdf}){{width={width}}}")
````

````python
# | include: true
render_pmx(
    width="70%",
    caption="One instrument, two score lines",
    source=r"""
1 1 4 4 4 4 0 0
1 2 16 0

t
./
c4 d e f /
g a b c /   
c b a g /
f e d c /    
/
""",
)
````

````python
# | include: true

render_pmx(
    caption="Two instruments, treble + treble, D major",
    width="55%",
    # After the header, one blank line per instrument - two here.
    # Clef string `tt` lists staves bottom-up; line ends at first space.
    source=r"""
2 2 3 4 3 4 0 2
1 1 16 0


tt
./
c4 d e /
c45 d e /
g4 a g /
g45 a g /
/
""",
)
````

````python
render_pmx(
    caption="Piano grand staff (one instrument, two staves)",
    width="55%",
    # Clef string `bt` = bass on the lower staff, treble on the upper
    # One blank line - both staves belong to the same instrument.
    source=r"""
2 1 4 4 4 4 0 0
1 1 16 0

bt
./
c43 d e f /
c45 d e f /
g4 a g f /
g4 a g f /
/
""",
)
````

**Below you can find an explanation of the \techword{PMX} header numbers.**

```
1 1 4 4 4 4 0 0 ← line 1
1  - number of voices
1  - number of instruments
4  - logical meter numerator (beats per measure)
4  - logical meter denominator (note value of the beat)
4  - printed meter numerator
4  - printed meter denominator
0  - pickup-bar length, in beats (0 = no pickup)
0  - key signature (+n = n sharps, -n = n flats, 0 = C major / A minor)

1 1 20 0 ← line 2
1  - number of pages (0 = let PMX decide)
1  - systems per piece (if npages>0); else average bars per system
20 - staff size in points (16, 20, 24, or 29; 20 is the default)
0  - indent of the first system, as a fraction of the line width (0 = none)
t  - active clef (treble); one letter per staff, listed bottom-up
./ - start of the note stream
```

A few things to keep in mind:

- *Note names.* \techword{PMX} uses `a`--`g` (Anglo-Saxon), `b` = B-natural. The Polish `H`/`B` convention is **not** supported -- transcribe `H` $\to$ `b`, Polish `B` $\to$ `bf`.
- *Multi-instrument.* Raise `voices`/`instruments` on line 1 and add one blank line per instrument before the clef string; for a piano grand staff use `instruments < voices`.
- *Common pitfalls.* Missing trailing `/` at bar ends; \techword{PMX} comments use `%` in column 1, not `#`; leftover `*.mx1`/`*.mx2` files should sit in `.gitignore`.

**Further reading.** The official manuals are the authoritative reference but tend to be&nbsp;terse; the community resources below are often a friendlier starting point:

- [\techword{PMX} package on CTAN](https://ctan.org/pkg/pmx) -- includes the full PMX user guide PDF by Don Simons.
- [\techword{MusiX\TeX} package on CTAN](https://ctan.org/pkg/musixtex) -- the underlying typesetter; consult its manual for \techword{\TeX}-level customisation (`\setname`, custom clefs, etc.).

## Electronical circuit schematics with \techword{schemdraw}
Circuit schematic could be generated automatically from code, using the [Schemdraw](https://schemdraw.readthedocs.io/en/stable/).

````python
# | fig-cap: This is an example circuit.
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing():
    elm.style(
        elm.STYLE_IEC
    )  # this is important: switch to European standard
    elm.Resistor().label("5$\\Omega$")
    elm.Capacitor()
    elm.ResistorVar().down()
    elm.Inductor()
    elm.Schottky().left()
    elm.Diode(fill=True).up()
    elm.Breaker().left()
    elm.BatteryCell().up()
````
